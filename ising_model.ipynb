```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

import renormalization as rg
```

# Initialize a renormalization group object

```python
RG = rg.RenormalizationGroup(dimension=1)
```

# Renormalizaiton-group flow

```python
j = .1
h = 0.001

RG.flow(j, h, n=30)
```

# Calculate and plot densities

```python
h = 0
temperature, density = RG.densities(h)
```

```python
plt.style.use('classic')
fig = plt.figure(figsize=(8, 6))
fig.set_facecolor('none')

lw = 2.5
fs = 20
plt.rc('lines', linewidth=lw)
plt.rc('axes', linewidth=lw)

ax = plt.subplot(111)
ax.plot(temperature, density[::,1], label=f'H = {h}')
ax.tick_params(axis="both", direction="in", left=True,
               width=lw, length=4, labelsize=fs)
ax.set_xlabel(r"Temperature $\mathrm{1/J}$", fontsize=fs)
ax.set_ylabel(r"Interaction Energy Density $\mathrm{\langle s_i s_j \rangle}$", fontsize=fs)
ax.legend()
plt.tight_layout()
#plt.savefig("densities", dpi=300)
plt.show()
```

```python
plt.style.use('classic')
fig = plt.figure(figsize=(8, 6))
fig.set_facecolor('none')

lw = 2.5
fs = 20
plt.rc('lines', linewidth=lw)
plt.rc('axes', linewidth=lw)

ax = plt.subplot(111)
ax.plot(temperature, density[::,2], label=f'H = {h}')
ax.tick_params(axis="both", direction="in", left=True,
               width=lw, length=4, labelsize=fs)
ax.set_xlabel(r"Temperature $\mathrm{1/J}$", fontsize=fs)
ax.set_ylabel(r"Magnetization $\mathrm{\langle s_i + s_j \rangle}$", fontsize=fs)
ax.legend()
plt.tight_layout()
#plt.savefig("densities", dpi=300)
plt.show()
```
